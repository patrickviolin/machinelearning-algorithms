### Pegar Cotações do Yahoo Finance com Python

- pandas_datareader
- yfinance

```python
import yfinance as yf
```

```python
ativos = ["ITUB3.SA", "^BVSP", "VALE3.SA", "PETR3.SA"]

start_date = "2023-01-01"
end_date = "2023-06-30"
price_table = yf.download(tickers=ativos, start=start_date, end=end_date)
data = price_table.drop(columns=price_table.filter(regex='High|Low|Open|Volume').columns)

display(data)
```

```python
price_table_norm = data / data.iloc[0]
display(price_table_norm)
price_table_norm.plot()
```
